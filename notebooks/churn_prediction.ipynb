```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
import joblib

df = pd.read_csv(r"C:\Users\GAUTAM\Downloads\Telco-Customer-Churn.csv")

# print(df.head())
# print(df.info())
# print(df.describe())
# print(df.shape)
# print(df.columns)
# print(df.isnull().sum())



df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors= "coerce")

# print(df["TotalCharges"].head(10))
# print(df["TotalCharges"].dtype)

# df["TotalCharges"].isnull().sum()


#cleaning the dataset

df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors= "coerce") 
# df["TotalCharges"].isnull().sum() # = 11 
df["TotalCharges"] = df["TotalCharges"].fillna(df["TotalCharges"].median())

print(df.isnull().sum()) # = 0 
                    


df = df.drop("customerID", axis = 1 )
print(df.columns)


df.to_csv("../data/clean-churn-telco-dataset.csv",index =False)


#Encoding 

Le = LabelEncoder()

categorial_columns = df.select_dtypes(include=["object" , "string"]).columns


for column in categorial_columns:
    df[column] = Le.fit_transform(df[column])
    print(df.head())



# ---- seperating the X and Y feature and result 


x = df.drop("Churn", axis = 1)
y =  df["Churn"]

print(x.shape)
print(y.shape)


# ------ tarining the model 

x_train , x_test , y_train , y_test = train_test_split(x,y , test_size =  0.2 , random_state = 42)


model = LogisticRegression()
model.fit(x_train,y_train)
print(model)


y_pred = model.predict(x_test)

print(y_pred[:10])


 # ---- comparison ---- 

comparison = pd.DataFrame({
    "Actual": y_test.values,
    "Predicted": y_pred
   
})

print(comparison.head(5))



# -- model checking 
accuracy = accuracy_score(y_test, y_pred)

print("Accuracy:", accuracy)

print(confusion_matrix(y_test, y_pred))

print(classification_report(y_test, y_pred))



#saving the model 

joblib.dump(model, "../models/customer_churn_predictor.pkl")

print("Model saved successfully!")


```
